```python

log_dir = '/scratch/midway3/jshe/molecular-attention-bias/homo_lumo_U'
model_type = 'GraphPETransformer'
architecture = 'E128H8D8'
```

```python
for root, dirs, files in os.walk(log_dir):
    if not files or 'model.pt' not in files: continue
    *_, this_model_type, this_radial_function_type, this_architecture, run_id = root.split('/')
    if model_type != this_model_type: continue
    #if radial_function_type != this_radial_function_type: continue
    if architecture != this_architecture: continue

    model = models.get_model(
        model_type=model_type,
        n_tokens=6, 
        out_features=3, 
        E=128, H=8, D=8, 
        dropout=0.1, 
        positional_encoding_type='RandomWalkPE', 
        k=16
    )

    model_file = os.path.join(root, 'model.pt')
    state_dict = torch.load(model_file, map_location=torch.device('cpu'))
    model.load_state_dict(state_dict)

    module = model.positional_encoding

    print(sum(p.numel() for p in module.parameters() if p.requires_grad))
```
