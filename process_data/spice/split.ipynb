```python
import numpy as np
import os
import json
```

```python
indir = '/scratch/midway3/jshe/data/spice/raw/'
outdir = '/scratch/midway3/jshe/data/spice/'
```

```python
datas = {fname.split('.')[0] : np.load(indir + fname) for fname in os.listdir(indir)}
```

```python
sample_indices = np.arange(datas['atomic_numbers'].shape[0])
np.random.shuffle(sample_indices)

train_indices = sample_indices[:int(len(sample_indices) * 0.8)]
val_indices = sample_indices[int(len(sample_indices) * 0.8):int(len(sample_indices) * 0.9)]
test_indices = sample_indices[int(len(sample_indices) * 0.9):]

print(f'Train: {len(train_indices)}')
print(f'Validation: {len(val_indices)}')
print(f'Test: {len(test_indices)}')
```

```python
y_mean = datas['y'][train_indices].mean(axis=0)
y_std = datas['y'][train_indices].std(axis=0)
```

```python
os.makedirs(os.path.join(outdir, 'train'), exist_ok=True)
os.makedirs(os.path.join(outdir, 'validation'), exist_ok=True)
os.makedirs(os.path.join(outdir, 'test'), exist_ok=True)

for name, data in datas.items():
    if name == 'y_labels': continue
    np.save(os.path.join(outdir, 'train', name + '.npy'), data[train_indices])
    np.save(os.path.join(outdir, 'validation', name + '.npy'), data[val_indices])
    np.save(os.path.join(outdir, 'test', name + '.npy'), data[test_indices])

for split in ['train/', 'validation/', 'test/']:
    np.save(os.path.join(outdir, split + 'y_labels.npy'), datas['y_labels'])
    np.save(os.path.join(outdir, split + 'y_mean.npy'), y_mean)
    np.save(os.path.join(outdir, split + 'y_std.npy'), y_std)
```
